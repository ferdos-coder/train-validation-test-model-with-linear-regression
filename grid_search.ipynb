```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\machine learning excercise\pista\جلسه 24\Advertising.csv")
```

```python
df
```

```python
#create X and y

X = df.drop('sales',axis=1)
y = df['sales']
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)

#scale Data

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_net_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],'l1_ratio':[0.1,0.5,0.7,0.95,0.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,param_grid=param_grid,scoring='neg_mean_squared_error',cv=5,verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_final = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_final)
```
